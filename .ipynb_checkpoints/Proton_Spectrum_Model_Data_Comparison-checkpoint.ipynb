## PAMELA data from Adriani et al., Science 332, 69 (2011)
### Data sets columns:
#### [0] Quantity $\quad$ [1] $<R>$ (GV) $\quad$ [2] Bin range $\quad$ [3] Flux (per R bin)  (GV m2 s sr)-1 $\quad$ [4] Stat Err $\quad$ [5] Syst Err

### * *More recent data in Adriani et al., ApJ 765, 91 (2013)*

```python
Z = 1
data_PAMELA_H = np.genfromtxt("Data/PAMELA_H_data.txt")
R_PAMELA = data_PAMELA_H[:,1]
E_PAMELA = R_PAMELA * Z
flux_R_PAMELA = data_PAMELA_H[:,3]
flux_PAMELA = flux_R_PAMELA / Z
error_PAMELA_stat = data_PAMELA_H[:,4]/Z; error_PAMELA_syst = data_PAMELA_H[:,5]/Z
error_PAMELA = np.sqrt(error_PAMELA_stat**2 + error_PAMELA_syst**2)
```

## Model by (Lipari $\&$ Vernetto, 2019) for hardening and softening of proton spectrum

```python
def Phi_twobreaks(E, K, E0, alpha1, alpha2, alpha3, omega1, omega2, Eb1, Eb2):
    delta1 = alpha2 - alpha1
    delta2 = alpha3 - alpha2
    return K * (E/E0)**(-alpha1) * (1 + (E/Eb1)**(1/omega1))**(-delta1 * omega1) * (1 + (E/Eb2)**(1/omega2))**(-delta2 * omega2)
```

```python
def Phi_onebreak(E, K, E0, alpha1, alpha2, omega1, Eb1):
    delta1 = alpha2 - alpha1
    phi = K * (E/E0)**(-alpha1) * (1 + (E/Eb1)**(1/omega1))**(-delta1 * omega1)
    return phi
```

```python
#PAMELA

alpha1 = 2.85
alpha2 = 2.67
Eb1 = 0.232 * 1e3 #GeV
omega1 = 0.25
K = 4.60 * 10**(-2) #(GeV m2 s sr)-1
E0 = 0.1 * 1e3 #GeV

Phi_PAMELA = Phi_onebreak(E_PAMELA, K, E0, alpha1, alpha2, omega1, Eb1)
```

```python
#AMS02

alpha1 = 2.849
alpha2 = 2.716
Eb1 = 0.336 * 1e3 #GeV
omega1 = 0.04
K = 4.42 * 10**(-2) #(GeV m2 s sr)-1
E0 = 0.1 * 1e3 #GeV

Phi_AMS02 = Phi_onebreak(E_AMS02, K, E0, alpha1, alpha2, omega1, Eb1)
```

```python
#DAMPE

alpha1 = 2.75
alpha2 = 2.58
alpha3 = 2.86
Eb1 = 0.574 * 1e3 #GeV
omega1 = 0.35
Eb2 = 12.8 * 1e3 #GeV
omega2 = 0.37
K = 1.15 * 10**(-5) #(GeV m2 s sr)-1
E0 = 2 * 1e3 #GeV

Phi_DAMPE = Phi_twobreaks(E_DAMPE, K, E0, alpha1, alpha2, alpha3, omega1, omega2, Eb1, Eb2)

# Real normalisation at 0.1 TeV
Phi_twobreaks(0.1 * 1e3, K, E0, alpha1, alpha2, alpha3, omega1, omega2, Eb1, Eb2)
```

```python
#CALET

alpha1 = 2.81
alpha2 = 2.55
Eb1 = 0.605 * 1e3 #GeV
omega1 = 0.25
K = 4.34 * 10**(-2) #(GeV m2 s sr)-1
E0 = 0.1 * 1e3 #GeV

Phi_CALET = Phi_onebreak(E_CALET, K, E0, alpha1, alpha2, omega1, Eb1)
```

```python
#CREAM

alpha2 = 2.58
alpha3 = 2.84
Eb2 = 16 * 1e3 #GeV
omega2 = 0.5
K = 1.22 * 10**(-5) #(GeV m2 s sr)-1
E0 = 2 * 1e3 #GeV

Phi_CREAM = Phi_onebreak(E_CREAM, K, E0, alpha2, alpha3, omega2, Eb2)
```

```python
#ATIC

alpha1 = 2.79
alpha2 = 2.62
Eb1 = 0.275 * 1e3 #GeV
omega1 = 0.34
K = 4.4 * 10**(-2) #(GeV m2 s sr)-1
E0 = 0.1 * 1e3 #GeV

Phi_ATIC = Phi_onebreak(E_ATIC, K, E0, alpha1, alpha2, omega1, Eb1)
```

```python
#All detectors

logE = np.arange(0,6,0.1)
E = 10**logE

alpha1 = 2.80
alpha2 = 2.57
alpha3 = 2.87
Eb1 = 0.67 * 1e3
omega1 = 0.27
Eb2 = 16 * 1e3
omega2 = 0.35
K = 4.40 * 10**(-2)
E0 = 0.1 * 1e3

Phi_all = Phi_twobreaks(E, K, E0, alpha1, alpha2, alpha3, omega1, omega2, Eb1, Eb2)
```
