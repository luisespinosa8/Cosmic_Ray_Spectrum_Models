# Light component analysis

```python
from pylab import *
```

```python
m_p = 0.938 # Proton mass
```

```python
Z = 1 # Atomic number
A = 1 # Mass number
n = 1 # Number of nucleons in H

data_PAMELA_H = np.genfromtxt("Data/Proton/PAMELA_H_data.txt")
R_PAMELA = data_PAMELA_H[:,1]
E_PAMELA_H = (R_PAMELA * Z) - (A * m_p)
flux_R_PAMELA = data_PAMELA_H[:,3]
flux_PAMELA_H = flux_R_PAMELA * (R_PAMELA / ((Z * R_PAMELA) - (A * m_p))) # Normalize to Ek
error_PAMELA_stat = data_PAMELA_H[:,4] * (R_PAMELA / ((Z * R_PAMELA) - (A * m_p))) # Normalize to Ek
error_PAMELA_syst = data_PAMELA_H[:,5] * (R_PAMELA / ((Z * R_PAMELA) - (A * m_p))) # Normalize to Ek
error_PAMELA_H = np.sqrt(error_PAMELA_stat**2 + error_PAMELA_syst**2) 

R_PAMELA
```

```python
Z = 2 # Atomic number
A = 4 # Mass number
n = 4 # Number of nucleons in H

data_PAMELA_He = np.genfromtxt("Data/Helium/PAMELA_He_data.txt")
R_PAMELA = data_PAMELA_He[:,1]
E_PAMELA_He = (R_PAMELA * Z) - (A * m_p)
flux_R_PAMELA = data_PAMELA_He[:,3]
flux_PAMELA_H = flux_R_PAMELA * (R_PAMELA / ((Z * R_PAMELA) - (A * m_p))) # Normalize to Ek
error_PAMELA_stat = data_PAMELA_He[:,4] * (R_PAMELA / ((Z * R_PAMELA) - (A * m_p))) # Normalize to Ek
error_PAMELA_syst = data_PAMELA_He[:,5] * (R_PAMELA / ((Z * R_PAMELA) - (A * m_p))) # Normalize to Ek
error_PAMELA_He = np.sqrt(error_PAMELA_stat**2 + error_PAMELA_syst**2) 

R_PAMELA
```
